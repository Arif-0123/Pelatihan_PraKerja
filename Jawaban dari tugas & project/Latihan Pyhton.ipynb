# Quiz Percabangan
Buatlah program untuk mendeteksi angka yang dimasukkan adalah :
1. Angka Nol
2. Genap Positif
3. Genap Negatif
4. Ganjil Positif
5. Ganjil Negatif

```python
#Quiz Latihan Percabangan

bil=int(input("Masukan angka :"))

if bil == 0:
    print("Angka Nol")
elif bil % 2 == 0:
    if bil > 0:
      print("Genap Positif")
    else:
      print("Genap Negatif")
else:
    if bil > 0:
      print("Ganjil Positif")
    else:
      print("Ganjil Negatif")

print("angka anda adalah ",bil)
```

```python
#Menggabungkan dua list menggunakan fungsi append

languages1=['Python', 'PHP', 'Java', ]
languages2=['C', 'C++', 'C#']
# Menambahkan List 2 ke List 1 dengan perintah Append (isinya tetap bagian kedua)
languages1.append(languages2)
print("Result: ", languages1)
print("Panjang Languages 1 :", len(languages1))
```

```python
#Menggabungkan dua list menggunakan fungsi extend
languages1=['Python', 'PHP', 'Java',]
languages2=['C', 'C++', 'C#']
# Menambahkan List 2 ke List 1 dengan perintah extend(isinya gabung)
languages1.extend(languages2)
print("Result: ",languages1)
print("Panjang Languages 1 :",len (languages1))
```

```python
#pandas adalah library python khusus untuk data analisis
#berikut cara import library pandas sekaligus bikin nama alias :pd
import pandas as pd
#Membuat tabel pandas baru bertipe dataframe dengan nama : df
#pd dipakai untuk membuat tabel pandas baru dari sebuah dictionary
dict={"A":[1,4,7],
      "B":[2,5,8],
      "C":[3,6,9]}

df = pd.DataFrame(dict)  #disini df adalah berupa tabel pandas
print(df)
df
```

```python
#Mengakses elemen Pandas df di atas.
# Pakai iloc[] maka cukup menyebutkan nomor index nya
print(df.iloc[0][0])

# Pakai Loc[] maka harus menyebutkan judul kolom dan baris
print(df.loc[0]['A'])

# Pakai at[] maka sama dengan Loc, tapi tanpa siku pemisah
print(df.at[0, 'A'])

# Pakai iat[] maka sama dengan iloc, tapi tanpa siku pemisah

print(df.iat [0,0])
```
